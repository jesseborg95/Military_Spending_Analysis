## Cleaning Process

Since multiple datasets are going to be used for this analysis, the best way to approach this is to initially read each CSV file while setting the country name as the index, which will be kept constant across all datasets. Then, the unwanted countries and years will be removed so that what is left is the relevant information, as mentioned above. To make it easier to concatenate the datasets, each individual one will be stacked according to the Country; this will make the data easy to read in one big dataframe and will also facilitate the plotting of data over a number of years for the different datasets.

Once this has been completed, the data can then be plotted in various ways to make it easy to observe changes over time, patterns or any correlations that may exist. Below, this cleaning process was carried out and then the resulting dataframe was used to plot the desired data.

```python
# Import the required tools from the Python library and then abbreviate them to make it easier and quicker to call upon when needed
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
# Import each dataset using the Pandas library, giving them appropriate names and setting the country name as the index
# Check to see if the data was read correctly for each data set by using the df.head() function
gdp = pd.read_csv('gdp.csv', index_col = 0)
gdp.head()
```

```python
percapita = pd.read_csv('percapita.csv', index_col = 0)
percapita.head()
```

```python
population = pd.read_csv('population.csv', index_col = 0)
population.head()
```

```python
spending = pd.read_csv('spending.csv', index_col = 0)
spending.head()
```

```python
# For the datasets, choose the desired countries and years using the loc function and then overwriting the variable as a new dataset
# Stack this new dataset, using level = 0 as a parameter to keep the country name as the main index
# Display the stacked dataset to check if it has been done successfully
gdp = gdp.loc[['United States', 'China', 'Russian Federation', 'Germany', 'United Kingdom', 'France', 'Italy', 'Saudi Arabia', 'Korea, Rep.', 'Australia'], '2013':'2018']
gdp = gdp.stack(level = 0)
display(gdp.head())

# This was done for all other datasets below to prepare for the merging of datasets
```

```python
percapita = percapita.loc[['United States', 'China', 'Russian Federation', 'Germany', 'United Kingdom', 'France', 'Italy', 'Saudi Arabia', 'Korea, Rep.', 'Australia'], '2013':'2018']
percapita = percapita.stack(level =  0)
display(percapita.head())
```

```python
population = population.loc[['United States', 'China', 'Russian Federation', 'Germany', 'United Kingdom', 'France', 'Italy', 'Saudi Arabia', 'Korea, Rep.', 'Australia'], '2013':'2018']
population = population.stack(level = 0)
display(population.head())
```

```python
spending = spending.loc[['United States', 'China', 'Russian Federation', 'Germany', 'United Kingdom', 'France', 'Italy', 'Saudi Arabia', 'Korea, Rep.', 'Australia'], '2013':'2018']
spending = spending.stack(level = 0)
display(spending.head())
```

```python
# To merge the datasets, use the pandas concat function and specify the datasets to join and the axis parameter set to 1 to join by columns
countries = pd.concat([gdp, percapita, population, spending], axis = 1)
display(countries.head()) # Check whether it has been done successfully
```

```python
# The different datasets have been merged as columns; with their names set as their column index. Rename the columns to the different dataset information using the - 
#  - rename function and specifying the column index as they have no name. 
countries.rename(columns={ countries.columns[0]: "GDP", countries.columns[1]:'Per Capita', countries.columns[2]:'Population', countries.columns[3]:'Military' }, inplace = True)

# Rename South Korea and Russia to make their names easier to type and reduce error
countries = countries.rename(index = {'Korea, Rep.':'South Korea', 'Russian Federation':'Russia'})
countries.loc[['South Korea', 'Russia'] , :] # Check to see if the renaming has worked
```

## Comparing the military data to that countryâ€™s GDP

```python
# Since the GDP values are much larger than military spending, they wont be comparable when plotted on the same graph
# To fix this, the values were scaled to make them easily comparable on the same graph
countries['GDP x (10^12)'] = countries['GDP'] / 1000000000000
countries['Military x (10^10)'] = countries['Military'] / 10000000000
countries.head()
```

```python
# Create a function to plot graphs for a specified country and dataset
# This function will be used to plot GDP against Military spending, so the title and axis labels are set accordingly
# Setting the line labels to the pertinent dataset to insert a legend
def plot1(country, dataset):
    data = countries.loc[country, dataset]
    plt.plot(data.index, data.values, label = dataset)
    plt.legend()
    plt.title('GDP Vs. Military Spending for ' + country)
    plt.xlabel('Year', fontsize = 12)
    plt.ylabel('Total ($)', fontsize = 12)
```

```python
# Create a for-loop using each country means that the plot will only need to be specified once, then they will be plotted automatically  
places = ['United States', 'China', 'Russia', 'Germany', 'United Kingdom', 'France', 'Italy', 'Saudi Arabia', 'South Korea', 'Australia']

for p in places:
    plt.figure(figsize = (8,8)) 
    plot1(p, 'GDP x (10^12)')
    plot1(p, 'Military x (10^10)')
    plt.show()
```

The graphs show the GDP and Military spending for the 10 different countries from the year 2013 to 2018.

Overall, the GDP of the countries seem to be going up, but in 2014 & 2015 the GDP of the countries except the US, South Korea and China stagnated or declined, before recovering in 2016. Experts say that this could be due to the fact that GDP is measured in USD and the value of the dollar increased significantly between 2014 & 2015. This is evident when looking at the GDP of the EU countries. Saudi Arabia's GDP barely changed over the 5 year period.

Over the same period, the military spending pattern is almost identical to that displayed by the GDP for most of the countries analysed. This makes sense because as a country's GDP increases, there is more money to spend on sectors such as military. Again, there are some special cases such as Australia where the military spending is almost exactly opposite as GDP or in Saudi Arabia where the military spending seems to be greater than most countries even though they have the lowest GDP. 

In general,  the military spending and GDP are very strongly correlated.

## Comparing the overall military spending of all 10 countries

```python
# Create a loop to plot a graph for absolute military spending for all 10 countries between 2013 and 2018
places = ['United States', 'China', 'Russia', 'Germany', 'United Kingdom', 'France', 'Italy', 'Saudi Arabia', 'South Korea', 'Australia']

plt.figure(figsize=(10,8))

for p in places:
    data = countries.loc[p, 'Military']
    plt.plot(data.index, data.values, label = p)
    plt.legend(loc = 'best')
    plt.title('Military spending from 2013 to 2018', fontsize = 14)
    plt.ylabel('Amount ($)', fontsize = 14)
    plt.xlabel('Year', fontsize = 14)
    
plt.show()
```

The graph above shows the overall military spending per year for each country between the years of 2013 and 2018.

From the graph, it is clear to see that the United States outspends every other country by a long shot, and China spends approximately double on military than the rest of the countries. It also shows that apart from the US and China, military spending per year has stagnated. Military spending in the US was going down until 2016, but after 2016; with a change of presidency, spending is once again on the rise.

It is expected for the US to outspend every other country by a mile because the US has many military bases in every continent and tend to consider themselves the world police. China on the other hand, is becoming a world superpower; meaning they have increased budgets, which is reflected on military spending. 

On the whole, countries dont seem to be spending more on military as the prospect of war is becoming less desirable and many countries have peace treaties with one another. This of course depends on who comes into power and what their views are.

```python
# Import mean from the statistics library to be able to calculate the average spending per year for each country between 2013 & 2018 inclusive
from statistics import mean

# Create a loop to plot a graph for average military spending per year for all 10 countries between 2013 and 2018
places = ['United States', 'China', 'Russia', 'Germany', 'United Kingdom', 'France', 'Italy', 'Saudi Arabia', 'South Korea', 'Australia']

plt.figure(figsize=(15,5))

# Since we are looking at 1 value per country, it is much easier to visualise with a bar plot
for p in places:
    data = countries.loc[p, 'Military']
    mean_spending = mean(data)
    plt.bar(p, mean_spending)
    plt.title('Average military spending per year between 2013 and 2018', fontsize = 14)
    plt.xlabel('Country', fontsize = 14)
    plt.ylabel('Amount ($)', fontsize = 14)
    
plt.show()
```

The bar plot shows the average military spending per year between 2013 and 2018 for the 10 different countries.

The plot confirms that the United States outspends every other country by at least 2.5 times on the military per year on average. It also shows how China is the second biggest military spender on average. Every other country is more or less on par with one another; with Russia and Saudi Arabia spending slightly more on average than the rest. Italy and Australia are spending the least.

## Comparing the military spending per person to the GDP per capita

```python
# To compare the military spending per person to the GDP per capita, create a new column for military spending per person; which is simply Military Spending / Population
countries['Military Per Person'] = countries['Military'] / countries['Population']

# Create columns which reduce the military spending per person and GDP per capita to comparable scales
countries['Military Per Person x (10^3)'] = round((countries['Military'] / countries['Population']) / 1000, 2)
countries['Per Capita x (10^5)'] = countries['Per Capita'] / 100000
countries.head()
```

```python
# Create a function to plot graphs for a specified country and dataset
# This function will be used to plot GDP per capita against military spending per person, so the title and axis labels are set accordingly
# Setting the line labels to the pertinent dataset, to insert a legend
def plot2(country, dataset):
    data = countries.loc[country, dataset]
    plt.plot(data.index, data.values, label = dataset)
    plt.legend()
    plt.title('GDP Per Capita Vs Military Spending Per Capita for ' + country)
    plt.xlabel('Year', fontsize = 12)
    plt.ylabel('Total ($)', fontsize = 12)
```

```python
# Create a for loop using each country means that the plot will only need to be specified once, then they will be plotted automatically  
places = ['United States', 'China', 'Russia', 'Germany', 'United Kingdom', 'France', 'Italy', 'Saudi Arabia', 'South Korea', 'Australia']

for p in places:
    plt.figure(figsize = (8,8))
    plot2(p, 'Per Capita x (10^5)')
    plot2(p, 'Military Per Person x (10^3)')
    plt.show()
```

The graphs show the GDP per capita and military spending per person for the 10 different countries from the year 2013 to 2018.

Since GDP and GDP per capita are linked, the GDP per capita of the countries has an almost identical pattern to that of the GDP; where in general, GDP per capita is going up except for 2014 & 2015 before recovering in 2016 for reasons previously mentioned. 

Over the same period, the military spending per capita follows a pattern almost identical to that displayed by the GDP per capita for most of the countries analysed. This makes sense because as a country's GDP per capita increases, there is more money to spend on sectors such as military. Again, there are some special cases such as Australia and Saudi Arabia where the military spending per capita has very little to no correlation with GDP per capita. 

In general,  the military spending per person and GDP per capita are strongly correlated.

```python
# To compare the military spending per person as a percentage of the GDP per capita, create a new column with the calculated percentage
countries['Military%'] = round((countries['Military Per Person'] / countries['Per Capita']) *100, 2)
countries.head()
```

```python
# Create a loop to plot a graph for military spending per person as a percentage of GDP per capita for all 10 countries between 2013 and 2018
places = ['United States', 'China', 'Russia', 'Germany', 'United Kingdom', 'France', 'Italy', 'Saudi Arabia', 'South Korea', 'Australia']

plt.figure(figsize=(12,10))

for p in places:
    data = countries.loc[p, 'Military%']
    plt.plot(data.index, data.values, label = p)
    plt.legend()
    plt.title('Military spending as a percentage of GDP from 2013 to 2018', fontsize = 14)
    plt.xlabel('Year', fontsize = 14)
    plt.ylabel('Percentage', fontsize = 14)
    
    
plt.show()
```

The graph above shows the military spending per person as a percentage of GDP per capita per year for each country between the years of 2013 and 2018.

From the graph, it is clear to see that there is completely different story when it comes to how much percent is spent on military. In this case, Saudi Arabia spends the greatest percentage on military by a long way. The United States, who outspends every other country in absolute value, actually spends less than 4% per year from their GDP per capita on military; which does not seem as outrageous. China, who was the second biggest spender, spends a similar percentage of their GDP per capita as the lowest spending countries: This is probably because of the sheer population of China.

In terms of percentage, Russia is the second biggest spender, but has been decreasing since 2016 and is almost on par with the US. The other countries; which spend the least amount on military, also spend the least percentage. 

On the whole, countries spend less than 5% of their GDP per capita on military. This is most likely because they are focusing their spending on other matters such as healthcare and the environment. This is also due to the fact that the prospect of war is becoming less desirable, but countries like Saudi Arabia who are in a war torn and hostile region, are forced to focus more of their budget on military spending.

```python
# Create a loop to plot a graph for average military spending per year as a percentage of GDP per capita for all 10 countries between 2013 and 2018
places = ['United States', 'China', 'Russia', 'Germany', 'United Kingdom', 'France', 'Italy', 'Saudi Arabia', 'South Korea', 'Australia']

# Since we are looking at 1 value per country, it is much easier to visualise with a bar plot
plt.figure(figsize=(15,5))

for p in places:
    data = countries.loc[p, 'Military%']
    mean_pct = mean(data)
    plt.bar(p, mean_pct)
    plt.title('Average military spending per year as a percentage of GDP between 2013 and 2018', fontsize = 14)
    plt.ylabel('Percentage', fontsize = 14)
    plt.xlabel('Country', fontsize = 14)
        
plt.show()
```

The bar plot shows the average military spending per year as a percentage of GDP per capita between 2013 and 2018 for the 10 different countries.

The plot shows just how much more Saudi Arabia spend as a percentage on military than every outher country. It also shows that the United States, even though it outspends every other country by at least 2.5 times on military spending per year on average, the percentage of which actually goes to military is quite normal and almost on par with the other countries. It also shows how China, depsite being the second biggest military spender on average, also dedicates a very small percentage on military.

The other countries follow the trend that absolute military spending is correlated with military spending as a percentage of GDP per capita.

## Fastest growing countries in military spending

```python
# Create a new dataframe with just the military spending data (absolute)
military = pd.DataFrame(countries['Military'])
military.head()
```

```python
# Calculate the absolute change and percentage change between years using the df.diff() and df.pct_change() functions
military['Change'] = military.diff()
military['Percentage Change'] = (military['Military'].pct_change())*100

# Since there is no data for 2012, drop 2013 for all countries as there will be no values
military = military.drop(index = '2013', level = 1)
military.head()
```

```python
# Create a loop to plot a graph for change in military spending per year for all 10 countries between 2014 and 2018
places = ['United States', 'China', 'Russia', 'Germany', 'United Kingdom', 'France', 'Italy', 'Saudi Arabia', 'South Korea', 'Australia']

plt.figure(figsize=(12,10))

for p in places:
    data = military.loc[p, 'Change']
    plt.plot(data.index, data.values, label = p)
    plt.legend()
    plt.title('Change in military spending by year from 2013 to 2018', fontsize = 14)
    plt.xlabel('Year', fontsize = 14)
    plt.ylabel('Amount ($)', fontsize = 14)
    
plt.show()
```

The graph shows the change in military spending each year between 2014 and 2018 for all 10 countries. 2013 was not included as there is no data for the change in that year.

The graph shows that the military spending for the majority of countries changes by less than 10 Billion USD per year.

The countries with the biggest changes are the US; whose military spending was decreasing, but began to increase after 2016. China, which overall has the biggest spending  over the past 5 years, with spending going up by around 20 billion each year since 2013. Russia's military spending is generally decreasing, as s Saudi Arabia's. 

From this graph, it shows there is a correlation that the countries which spend the most on military,  tend to make the biggest increases on military spending per year.

```python
# Create a loop to plot a bar graph for average change in military spending per year for all 10 countries between 2014 and 2018
places = ['United States', 'China', 'Russia', 'Germany', 'United Kingdom', 'France', 'Italy', 'Saudi Arabia', 'South Korea', 'Australia']

# Since we are looking at 1 value per country, it is much easier to visualise with a bar plot
plt.figure(figsize=(15,5))

for p in places:
    data = military.loc[p, 'Change']
    mean_c = mean(data)
    plt.bar(p, mean_c)
    plt.title('Average change in military spending per year between 2014 and 2018', fontsize = 14)
    plt.ylabel('Amount ($)', fontsize = 14)
    plt.xlabel('Country', fontsize = 14)
        
plt.show()
```

This bar plot shows the average change in military spending each year between 2014 and 2018 for all 10 countries. 2013 was not included as there is no data for the change in that year.

This plot shows that over the past 5 years, China's military has been growing by approximately 10x more than the other countries in absolute spending. The US's military spending is also growing, but at a much slower rate. The biggest surprise is the fact that out of the 10 countries, South Korea's army has been growing the third fastest on average in absolute spending over the past 5 years. 

Even though Saudi Arabia was spending the most percent of their GDP on military, their military spending has had almost no increase on average in the past 5 years. This is because Saudi Arabia spends a similar amount to many of the countries, despite their GDP being the lowest.

When it comes to average increase in military spending per year, China is the stand-out country as they spend on average, almost 15 Billion more each year on military than the previous.
